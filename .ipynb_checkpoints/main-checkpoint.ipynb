### Importing of packages

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from textwrap import wrap
```

### Declaration of read functions
This function are used throughout the notebook to extract data from the datasets.

```python
def read_data_from_csv(filename, dtype, delimiter=',', names=True):
    return np.genfromtxt(filename, dtype=dtype, delimiter=delimiter, names=names)
```

### Reading of datasets

```python
dtype = ['int32', 'U50', 'U50', 'int32']
accidents_dtype = ['int32', 'U128', 'U128', 'U128', 'int32']

# Annual motor vehicle population by vehicle type
annual_motor_vehicle_population_by_type_data = read_data_from_csv("./data/annual-motor-vehicle-population-by-vehicle-type.csv", dtype)

# Road traffic conditions data
road_traffic_conditions_data = read_data_from_csv("./data/road_traffic_condition.csv", ['int32', 'float', 'float', 'int32'])

# Accidents in Singapore
accidents_data = read_data_from_csv("./data/causes-of-accidents-by-severity-of-injury-sustained.csv", dtype=accidents_dtype)
```

### Function declarations
These functions are used throughout the notebook to perform data extraction, some of them extract data from the datasets and some of them are used to perform data analysis.

```python
# The function below is used to extract the data for a given type and category
def extract_data_by_type_or_category(t, c=''):
    if(len(c) == 0):
        return annual_motor_vehicle_population_by_type_data[annual_motor_vehicle_population_by_type_data['type'] == t]
    return annual_motor_vehicle_population_by_type_data[np.logical_and(annual_motor_vehicle_population_by_type_data['category'] == c, annual_motor_vehicle_population_by_type_data['type'] == t)]

# The function below is used to extract the data for a given year for a given type(s) of vehicles
def extract_data_by_year(year, types):
    extracted_data = {}
    for label in types:
        value_to_push = annual_motor_vehicle_population_by_type_data[np.logical_and(annual_motor_vehicle_population_by_type_data['year'] == year, annual_motor_vehicle_population_by_type_data['type'] == label)]['number']
        if(len(value_to_push) > 0):
            extracted_data[label] = value_to_push[0]

    return extracted_data.keys(), extracted_data.values()

# The function below is used to extract the data for a given year for a given type(s) of vehicles for a given year
def extract_fatal_and_injury_by_year(year):
    accidents_fatal = accidents_data[np.logical_and(accidents_data['accident_classification'] == 'FATAL', accidents_data['year'] == year)]
    accidents_injury = accidents_data[np.logical_and(accidents_data['accident_classification'] == 'INJURY', accidents_data['year'] == year)]

    accidents_fatal_sum = accidents_fatal['number_of_accidents'].sum()
    accidents_injury_sum = accidents_injury['number_of_accidents'].sum()

    return [accidents_fatal_sum, accidents_injury_sum]

```

### Text analysis of dataset #1 (annual-motor-vehicle-population-by-vehicle-type.csv)

```python
data_private_cars = extract_data_by_type_or_category('Private cars')
data_motorcycles = extract_data_by_type_or_category('Motorcycles and Scooters')
data_lgv = extract_data_by_type_or_category('Light Goods Vehicles (LGVs)')
```

```python
# Calculate min, max, mean, median, std and print them out
min_private_cars = np.min(data_private_cars['number'])
max_private_cars = np.max(data_private_cars['number'])
mean_private_cars = np.mean(data_private_cars['number'])
median_private_cars = np.median(data_private_cars['number'])
std_private_cars = np.std(data_private_cars['number'])

data_for_text_analysis = {}
unique_vehicle_types = np.unique(annual_motor_vehicle_population_by_type_data['type'])
unique_vehicle_category = np.unique(annual_motor_vehicle_population_by_type_data['category'])

for type in unique_vehicle_types:
    data = extract_data_by_type_or_category(type)

    numbers = {
        'min': np.min(data['number']),
        'max': np.max(data['number']),
        'mean': np.mean(data['number']),
        'median': np.median(data['number']),
        'std': np.std(data['number'])
    }

    data_for_text_analysis[type] = numbers


print("There are ", len(annual_motor_vehicle_population_by_type_data), " rows in the dataset.")
print("")

print("There are ", len(unique_vehicle_types), " unique vehicle types in the dataset.")
print("")

print("There are ", len(unique_vehicle_category), " unique vehicle categories in the dataset.")
print("")

for type in data_for_text_analysis:
    print("====" + type + "====")
    print("Min:\t", data_for_text_analysis[type]['min'])
    print("Max:\t", data_for_text_analysis[type]['max'])
    print("Mean:\t", data_for_text_analysis[type]['mean'])
    print("Median:\t", data_for_text_analysis[type]['median'])
    print("Std:\t", data_for_text_analysis[type]['std'])
    print("")
```

## Bar graph of the number of accidents in Singapore from 2012 to 2018
This bar graph shows the number of accidents in Singapore from 2012 to 2018.

```python


# Use numpy to generate 2012 to 2018
years = np.arange(2012, 2019)

labels = years
fatal_datapoints = []
injury_datapoints = []

# Loop through each years and extract the data
for year in years:
    fatal, injury = extract_fatal_and_injury_by_year(year)
    fatal_datapoints.append(fatal)
    injury_datapoints.append(injury)



# x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(1,2, figsize=(15, 5))
ax[0].bar(years, fatal_datapoints, width, label='Fatal')
ax[1].bar(years, injury_datapoints, width, label='Injury')

# Add some text for labels, title and custom x-axis tick labels, etc.

ax[0].set_title('Number of Fatal Accidents each year')
ax[1].set_title('Number of Non-Fatal Accidents each year')

ax[0].set_ylabel('Number of Accidents')
ax[0].set_xlabel('Year')

ax[1].set_ylabel('Number of Accidents')
ax[1].set_xlabel('Year')

# ax[0].set_xticks(x, labels)
# ax[1].set_xticks(x, labels)
# ax.legend()

# ax.bar_label(rects1, padding=3)
# ax.bar_label(rects2, padding=3)

plt.suptitle('Number of Accidents by Severity of Injury by year in Singapore')

plt.show()
```

#### Insights/Remarks from Bar Graph
-   The number of fatal accidents in have been somewhat on a decreasing trend overall from 2012.
-   The number of non-fatal accidents does not have a trend as it decreases from 2012 to 2013 and then increases from 2013 to 2014 before decreasing from then onwards till 2018. I guess the only trend that could be made is that from 2014, the number of non-fatal accidents have been decreasing.

#### Process
1.  Get the number of accidents by the year and the road user group which are either 'Drivers or Riders or Cyclists' or 'Pedestrians'. This is done by one of the functions declared earlier.
2.  Extract the years needed.
3.  Plot the histogram.

```python
def extract_number_of_accidents_by_year_and_road_user_group(year, road_user_group):
    d = accidents_data[np.logical_and(accidents_data['road_user_group'] == road_user_group, accidents_data['year'] == year)]['number_of_accidents']
    return d

num_of_accidents_2012_drivers = extract_number_of_accidents_by_year_and_road_user_group(2012, 'Drivers or Riders or Cyclists')
num_of_accidents_2015_drivers = extract_number_of_accidents_by_year_and_road_user_group(2015, 'Drivers or Riders or Cyclists')
num_of_accidents_2018_drivers = extract_number_of_accidents_by_year_and_road_user_group(2018, 'Drivers or Riders or Cyclists')

num_of_accidents_2012_pedestrians = extract_number_of_accidents_by_year_and_road_user_group(2012, 'Pedestrians')
num_of_accidents_2015_pedestrians = extract_number_of_accidents_by_year_and_road_user_group(2015, 'Pedestrians')
num_of_accidents_2018_pedestrians = extract_number_of_accidents_by_year_and_road_user_group(2018, 'Pedestrians')

fig, ax = plt.subplots(2,3, figsize=(20, 10))

bins = 5;

ax[0,0].set_title('Drivers, Riders or Cyclists - 2012')
ax[0,1].set_title('Drivers, Riders or Cyclists - 2015')
ax[0,2].set_title('Drivers, Riders or Cyclists - 2018')

ax[1,0].set_title('Pedestrians - 2012')
ax[1,1].set_title('Pedestrians - 2015')
ax[1,2].set_title('Pedestrians - 2018')

ax[0,0].grid()
ax[0,1].grid()
ax[0,2].grid()

ax[1,0].grid()
ax[1,1].grid()
ax[1,2].grid()

# Set frequency labels for every one of them
ax[0,0].set_ylabel('Frequency')
ax[0,1].set_ylabel('Frequency')
ax[0,2].set_ylabel('Frequency')
ax[1,0].set_ylabel('Frequency')
ax[1,1].set_ylabel('Frequency')
ax[1,2].set_ylabel('Frequency')

ax[0,0].set_xlabel('Number of Accidents')
ax[0,1].set_xlabel('Number of Accidents')
ax[0,2].set_xlabel('Number of Accidents')

ax[1,0].set_xlabel('Number of Accidents')
ax[1,1].set_xlabel('Number of Accidents')
ax[1,2].set_xlabel('Number of Accidents')

ax[0,0].hist(num_of_accidents_2012_drivers, bins=bins, color='red')
ax[0,1].hist(num_of_accidents_2015_drivers, bins=bins, color='green')
ax[0,2].hist(num_of_accidents_2018_drivers, bins=bins, color='blue')

ax[1,0].hist(num_of_accidents_2012_pedestrians, bins=bins, color='red')
ax[1,1].hist(num_of_accidents_2015_pedestrians, bins=bins, color='green')
ax[1,2].hist(num_of_accidents_2018_pedestrians, bins=bins, color='blue')

plt.suptitle('Number of Accidents by Drivers or Riders or Cyclists AND Pedestrians in 2012, 2015 and 2018 in Singapore')

# print(accidents_data[accidents_data['road_user_group'] == 'Drivers or Riders or Cyclists']['number_of_accidents'])
```

## Histogram of causes of accidents caused by Drivers, Riders or Cyclists from 2012 to 2018.

```python
# Extract the unique causes of accidents for "Drivers or Riders or Cyclists"
accidents_data_drivers = accidents_data[accidents_data['road_user_group'] == 'Drivers or Riders or Cyclists']

unique_causes_drivers = np.unique(accidents_data_drivers['causes_of_accident'])

fig, ax = plt.subplots(3, math.ceil(len(unique_causes_drivers)/3), figsize=(20, 10))

bins = 10

# Add space in each height
plt.subplots_adjust(hspace=0.8, wspace=0.5)

plt.suptitle('Frequency of Accidents by Cause of Accidents for Drivers or Riders or Cyclists in Singapore between 2012 and 2018')

for index, cause in enumerate(unique_causes_drivers[:4]):
    ax[0][index].hist(accidents_data_drivers[accidents_data_drivers['causes_of_accident'] == cause]['number_of_accidents'], bins=bins, label=cause)
    ax[0][index].set_title('\n'.join(wrap(cause,30)))
    ax[0][index].set_xlabel('Number of Accidents')
    ax[0][index].set_ylabel('Frequency')

for index, cause in enumerate(unique_causes_drivers[4:8]):
    ax[1][index].hist(accidents_data_drivers[accidents_data_drivers['causes_of_accident'] == cause]['number_of_accidents'], bins=bins, label=cause)
    ax[1][index].set_title('\n'.join(wrap(cause,30)))
    ax[1][index].set_xlabel('Number of Accidents')
    ax[1][index].set_ylabel('Frequency')

for index, cause in enumerate(unique_causes_drivers[8:]):
    ax[2][index].hist(accidents_data_drivers[accidents_data_drivers['causes_of_accident'] == cause]['number_of_accidents'], bins=bins, label=cause)

    ax[2][index].set_title('\n'.join(wrap(cause,30)))
    ax[2][index].set_xlabel('Number of Accidents')
    ax[2][index].set_ylabel('Frequency')
```

#### Insights/Remarks from Histogram
-   The graph is skewed to the right or positively skewed for majority of them.
-   There is no 'definite' trend for accidents that happened between 2012 to 2018.
-   Note that for the graph "Failing to Keep a Proper Lookout". The bins are not the same as the other graphs. This is because the number of accidents for this cause is much higher than the other causes. Likewise for all graphs. The point of the histogram is to see the frequency of the causes of accidents and not the number of accidents.
